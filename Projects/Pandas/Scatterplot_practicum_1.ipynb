```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x=df['size'], y=df['price'])
plt.show()
```

```python
df = df[df['size']<250]
```

```python
sns.scatterplot(data=df, x=df['size'], y=df['price'])
plt.show()
```

```python
sns.regplot(data=df, x=df['size'], y=df['price'])
plt.show()
```

```python
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['rooms'])
plt.show()
```

```python
sns.scatterplot(data=df, x=df['size'], y=df['price'], hue=df['district'])
plt.show()
```

```python
sns.swarmplot(data=df, x=df['district'], y=df['price'])
plt.xticks(rotation=90)
plt.show()
```
