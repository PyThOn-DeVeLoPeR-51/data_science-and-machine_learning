```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
sns.histplot(data=df, x=df['price'])
plt.show()
```

```python
df = df[df['price']<200000]
```

```python
sns.histplot(data = df, x=df['price'])
plt.show()
```

```python
sns.histplot(data=df, x=df['size'])
plt.show()
```

```python
sns.histplot(data=df, x=df['rooms'])
plt.show()
```

```python
df = df[df['rooms']<7]
```

```python
sns.displot(data=df, x=df['price'], hue=df['rooms'], aspect=2, palette="Accent")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x=df['price'], hue=df['rooms'], multiple="stack", aspect=2, palette="Accent")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df[df['rooms']<4], x='price', hue='rooms', aspect=2, multiple='stack', palette='Reds')
plt.show()
```

```python
sns.displot(data=df[df['rooms']>3], x='price', col='rooms')
plt.show()
```

```python
sns.displot(data=df, kind='kde', x='size', aspect=2)
plt.show()
```

```python
sns.displot(data=df, kde=True, x='size', aspect=1.5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df[df['rooms'].isin([2,3,5])], x='price', hue='rooms', shade=True, palette="Accent", multiple='stack')
plt.show()
```

```python
sns.displot(data=df, x='size', kind='ecdf', aspect=1.5)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, kind='ecdf', x='price', aspect=1.5)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
