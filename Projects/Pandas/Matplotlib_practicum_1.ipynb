```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import ssl
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
data = pd.read_html("https://en.tutiempo.net/climate/ws-384570.html")
data[0]
```

```python
df = pd.DataFrame(data[0])
df.head()
```

```python
df.columns.values
```

```python
df = df[df['Year']>1990]
df.head()
```

```python
 df.set_index('Year', inplace=True)
 df.head()
```

```python
df.info()
```

```python
df = df.replace("-", np.nan)
df.head()
```

```python
df = df.astype(dtype='float')
df.info()
```

```python
df.head(10)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.drop('PP', axis=1, inplace=True)
df.head()
```

```python
df = df.fillna(method = "ffill")
df.head()
```

```python
df.to_csv("t_harorat.csv")
```

```python
# plt.figure(figsize=(10,5))
x = df['T']
plt.title("Avarenge Temperature")
plt.plot(x)
plt.show()
```

```python
# plt.figure(figsize=(18, 9))
x = df["T"]
plt.title("Average Temperature")
plt.xlabel('Years')
plt.ylabel("Temp")
plt.grid()
plt.plot(x, linewidth=2, linestyle="--", color="g", marker="h", markerfacecolor="r")
plt.show()
```

```python
minT = df["Tm"]
maxT = df["TM"]
# plt.figure(figsize=(20, 10))
plt.title('Average Temperature of Tashkent')
plt.xlabel("Year")
plt.ylabel("Temp")
plt.grid()
plt.legend(["Average minimum temperature", "Average maximum temperature"], loc=2, fontsize="x-large")
plt.plot(minT, linewidth=2, linestyle="--", marker="h", markerfacecolor="g")
plt.plot(maxT, linewidth=2, linestyle="-", marker="v", markerfacecolor="r")
plt.savefig('tempMinMax.png')
plt.show()
```
