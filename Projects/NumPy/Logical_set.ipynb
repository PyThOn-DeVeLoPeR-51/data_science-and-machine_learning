```python
import numpy as np
```

```python
xarr = np.arange(1.1, 1.6, 0.1)
xarr
```

```python
yarr = np.arange(2.1, 2.6, 0.1)
yarr
```

```python
cond = np.array([True, False, True, True, False])
```

```python
# where logical set function 1-exercise
result = np.where(cond, xarr, yarr)
result
```
