```python
import numpy as np
```

```python
# UNARY FUNCTIONS
```

```python
arr = np.arange(10)
arr
```

```python
np.sqrt(arr) # sqrt function
```

```python
np.square(arr) # square function
```

```python
np.exp(arr) # exp function
```

```python
np.log2(arr[1:]) # log function
```

```python
arr1 = np.random.randn(6)
arr1
```

```python
qoldiq, butun = np.modf(arr1) # modf function
```

```python
print(qoldiq)
print(butun)
```

```python
np.sign(arr1) # sign function
```

```python
arr1[0] = np.NaN
arr1
```

```python
np.isnan(arr1) # isnan function
```
