```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/china_gdp.csv")
df.head()
```

```python
x_data, y_data = (df['Year'].values, df['Value'].values)
plt.plot(x_data, y_data, 'ro')
plt.show()
```

```python
x = np.asanyarray(df[['Year']]/df.Year.max())
y = np.asanyarray(df[['Value']]/df.Value.max())
```

```python
PLN_model = LinearRegression()
```

```python
poly_features = PolynomialFeatures(degree=5, include_bias=False)
x_poly = poly_features.fit_transform(x)
```

```python
PLN_model.fit(x_poly, y)
```

```python
y_predict = PLN_model.predict(x_poly)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y, 'ro', label='data')
plt.plot(y_predict, linewidth=4, label='fit')
plt.show()
```

```python
print("MAE:", mean_absolute_error(y_predict,y))
```
