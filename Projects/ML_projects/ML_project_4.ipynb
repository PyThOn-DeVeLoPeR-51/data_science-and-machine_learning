```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head()
```

```python
housing = df[df['district'] == 'Чиланзарский']
housing.head()
```

```python
train_set, test_set = train_test_split(housing, test_size=0.10, random_state=42)
```

```python
MLR_model = LinearRegression()
```

```python
x_train = np.asanyarray(train_set[['rooms', 'size', 'level', 'max_levels']])
y_train = np.asanyarray(train_set[['price']])
MLR_model.fit(x_train, y_train)
print(f"'Coeffitcients:' {MLR_model.coef_}")
print(f"'theta0:' {MLR_model.intercept_}")
```

```python
x_train
```

```python
sns.pairplot(housing[['rooms', 'size', 'level', 'max_levels', 'price']])
plt.show()
```

```python
x_test = np.asanyarray(test_set[['rooms', 'size', 'level', 'max_levels']])
y_test = np.asanyarray(test_set[['price']])

y_predicted = MLR_model.predict(x_test)
```

```python
MAE = mean_absolute_error(y_test, y_predicted)
RMSE = np.sqrt(mean_squared_error(y_test, y_predicted))
print(f"MAE = {MAE}")
print(f"RMSE = {RMSE}")
```
